## class 를 완성하기 위해, random module 을 사용해야한다. 임의의 단어를 선택하기 위함이다. 
## 이 클래스의 주요 특징은 아래와 같다.

## 코드에서 확인할 수 있듯이, WORDS 리스트 중에 임의로 하나의 단어를 선택하고, 해당 단어를 Hangman 클래스에 넣어 hangman 이라는 클래스 인스턴스를 만든 후, 인스턴스의 play() 메소드를 호출하여 게임을 진행한다. 

````python
import random # 랜덤 모듈을 가져온다 


HANGMAN = [
    '________',
    '|       |',
    '|       O',
    '|       |',
    '|      /|\ ',
    '|       |',
    '|      / \ '
]

WORDS = ['casa' , 'car' , 'mono' , 'elevator' , 'python' , 'java' ,'pleasure' , 'young' , 'festival' , 'sing' , 'class'] # 사용할 단어들을 리스트로 가져온다

class Hangman():

    def __init__(self, word_to_guess):
        self.failed_attempts = 0 # 실패 횟수를 0으로 초기화
        self.word_to_guess = word_to_guess
        self.game_progress = list('_' * len(self.word_to_guess)) # 랜덤으로 선택된 단어의 길이만큼 밑줄 형성

    def find_indexes(self, letter):
        
        return [i for i, char in enumerate(self.word_to_guess) if letter == char]

    def is_invalid_letter(self, input_):
        
        return input_.isdigit() or (input_.isalpha() and len(input_) > 1)

    def print_game_status(self):
        
        print('\n')
        print('\n'.join(HANGMAN[:self.failed_attempts])) #join함수를 이용해서 현재 그림을 보여준다
        print('\n')
        print(' '.join(self.game_progress))

    def update_progress(self, letter, indexes):
        
        for index in indexes:
            self.game_progress[index] = letter

    def get_user_input(self):
        user_input = input('\n Please type a letter: ')
        return user_input

    def play(self):
        
        while True:
            self.print_game_status()
            user_input = self.get_user_input()

            if self.is_invalid_letter(user_input):
                print('The input is not a letter!')
                continue

            if user_input in self.game_progress:
                print('\n 이미 입력한 단어 입니다.')
                continue

            if user_input in self.word_to_guess: # 유저가 단어를 맞췄을 때 실행되는 조건문
                indexes = self.find_indexes(user_input)
                self.update_progress(user_input, indexes)
                
                if self.game_progress.count('_') == 0: # self.game_progress 리스트에 '_'이 없으면 즉 유저가 단어들을 다 맞췄을 때 게임이 종료되는 조건문
                    print('\n¡Yay! You win!')
                    print('The word is: {0}'.format(self.word_to_guess))
                    quit()
            else:
                self.failed_attempts += 1
        print("\n You lost!")

                                        # ----------------------- 게임 실행 --------------------------- #


if __name__ == '__main__':
    word_to_guess = random.choice(WORDS)
    hangman = Hangman(word_to_guess)
    hangman.play()
````
